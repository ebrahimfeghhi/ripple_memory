```python
def z_score(power):
    # mean center by mean across time and trials 
    # then divide by standard deviation of the average across timesteps
    power = (power - np.mean(power)) / np.std(np.mean(power, axis=1),axis=0)
    return power

def process_power(power, sess_dat):
    
    power_ds = decimate(np.log10(power), 10)
    
    ds_z = np.zeros_like(power_ds.squeeze())
    for sess in np.unique(sess_dat):
        sess_idxs = np.argwhere(sess_dat == sess).squeeze()
        ds_z[sess_idxs] = z_score(power_ds[sess_idxs])
              
    return ds_z
```

```python
elec_high_theta_frac = []
elec_high_theta_frac_dict = {}
for subj in np.unique(dd_trials_updated['subj']):
    subj_idxs = np.argwhere(dd_trials_updated['subj']==subj)
    for elec in np.unique(dd_trials_updated['elec_names']):
        elec_idxs = np.argwhere(dd_trials_updated['elec_names']==elec)
        subj_elec_idxs = np.intersect1d(subj_idxs, elec_idxs)
        if len(subj_elec_idxs) > 0:
            frac_high_theta = np.intersect1d(high_theta_power_trials, subj_elec_idxs).shape[0]/subj_elec_idxs.shape[0]
            elec_high_theta_frac.append(frac_high_theta)
            elec_high_theta_frac_dict[f'{subj}_{elec}'] = float(frac_high_theta)
```
