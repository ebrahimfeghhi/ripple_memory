```python
def return_trial_count_per_subject(dd_trials, catFR, organize_by): 
    
    organize_by_labels = dd_trials[organize_by]

    if catFR:
        num_trials = 'num_trials_cat'
    else:
        num_trials = 'num_trials_FR'

    num_trials_o_arr = []
    organize_name = []
    for o in np.unique(organize_by_labels):
        
        num_trials_o = np.argwhere(organize_by_labels==o).shape[0]
        
        organize_name.append(o)
        num_trials_o_arr.append(num_trials_o)

    trial_count_info = pd.DataFrame({organize_by: organize_name, num_trials:num_trials_o_arr})
    
    return trial_count_info
```

```python
def compute_fraction_clustered(clust_int_subj,  electrode_count, thresh=10):
    
    total_trials = clust_int_subj.shape[0]
    # a word follows, and it is semantically and/or temporally related
    num_clustered = np.argwhere(clust_int_subj > 1).shape[0]
    # a word follows, but not semantically or temporally related 
    num_not_clustered = np.argwhere(clust_int_subj == -1).shape[0] 
    
    if num_clustered/electrode_count < thresh or num_not_clustered//electrode_count < thresh:
        print("Too little data")
        s_i = None
    else:       
        s_i = num_clustered/(num_clustered+num_not_clustered)
    return s_i
```

Let's now try doing things at the list level. So basically, take a given list,  
and first store the number of items correctly recalled in that list (C) and its semantic structure.  
Next, randomly sample C items from a list with that structure N times (N=1000). 
Once that's done, 

```python
def convert_cat_to_ints(category_array):
    
    '''
    :param list category: category for each item in the presented list
    
    Converts category_array into an integer array, such that each category corresponds
    to a unique category. 
    '''
    
    unique_strings = np.unique(category_array)

    # Create a dictionary to map each unique string to an integer
    string_to_int = {string: idx for idx, string in enumerate(unique_strings)}

    # Replace each string in the array with its corresponding integer using the mapping
    integer_cat_array = np.vectorize(string_to_int.get)(category_array)

    return list(integer_cat_array)

def compute_available_lags(prev_rp, available_positions):
    
    '''
    Computes all the available lags given the previously recalled item (prev_rp)
    and the list of item positions that are available for recall (i.e. have not
    already been recalled) (available_positions). 
    
    '''
    
    available_lags = []
    for p in available_positions:
        lag = np.abs(prev_rp - p)
        available_lags.append(lag)
            
    return np.array(available_lags)

def compute_temporal_factor_score(recalled_position, list_length=12):
    
    '''
    :param list recalled_position: list of shape num_recalled items, where 
    each element corresponds to the index of the item during encoding, and the 
    order of the elements corresponds to the recall order.
    :param list_length int: number of items presented in the list
    
    Computes the temporal clustering score as defined by Kragel et. al 2021. 
    For each item, the list of available temporal lags is computed and compared 
    to the next item that was recalled. A score of 1 indicates the item that 
    was the most temporally proximal was recalled, and a score of 0 indicates the furthest
    temporal item was recalled.  
    '''
    
    available_positions = [i for i in range(1,list_length+1,1)]
    prev_rp = recalled_position[0]
    available_positions.remove(prev_rp)
    
    percentile_arr = []

    for rp in recalled_position[1:]:
        
        lag = np.abs(rp-prev_rp) # compute lag 
    
        # a sorted list of all possible lags 
        available_lags = compute_available_lags(prev_rp, available_positions)
        
        num_lags = available_lags.shape[0]
        
        # if there is only one lag distance you can choose, there is nothing to pick
        # from 
        if len(np.unique(available_lags)) == 1:
            continue
        
        # the first term gives the fraction of available lags that are stricly greater than the observed lag
        # the second term give1s the fraction of available lags that are equal to or greater than the observed lag
        # by taking the mean, we are saying the observed lag is in the middle of the pile of the available lags 
        # that are equal to it 
        score = ((available_lags > lag).mean() + (available_lags >= lag).mean() ) / 2

        percentile_arr.append(score)
        
        prev_rp = rp
        
        try:
            available_positions.remove(rp) # remove item
        except:
            print(rp, available_positions)

    return np.mean(percentile_arr)

def compute_available_semantic_dists(available_cats, prev_cat):
    
    available_semantic_dists = []
    for c in available_cats:
        
        if c == prev_cat:
            available_semantic_dists.append(1)
        else:
            available_semantic_dists.append(0)
            
    return np.array(available_semantic_dists)
    

def compute_semantic_factor_score(recalled_position, category_array, list_length=12):

    '''
    :param list recalled_position: list of shape num_recalled items, where 
    each element corresponds to the index of the item during encoding, and the 
    order of the elements corresponds to the recall order.
    
    :param list category array: list of length list_length, 
    where each list element is assigned an integer corresponding its category. 
    
    :param list_length int: number of items presented in the list
    
    Computes the semantic clustering score based on Kragel et. al 2021
    '''
    
    # category of the item recalled at the first position
    prev_cat = category_array[recalled_position[0]-1]
 
    semantic_score = []
    
    available_cats = deepcopy(category_array)
    
    # .remove just removes the first instance of the category (not all) 
    available_cats.remove(prev_cat)
    
    for rp in recalled_position[1:]:
        
        # get category of next item
        try:
            cat = category_array[rp-1]
        except:
            pass
        
        # there are no more words from that category, so skip
        # because it's not possible to semantically cluster at this point
        if prev_cat not in available_cats:
            continue
            
        # score of 1 if from the same category, 
        # else 0 
        if cat == prev_cat:
            semantic_dist = 1
        else:
            semantic_dist = 0 
            
        # compute the distribution of available semantic transitions
        available_semantic_dists = compute_available_semantic_dists(available_cats, prev_cat)
        
        score = ((available_semantic_dists < semantic_dist).mean() + (available_semantic_dists <= semantic_dist).mean() ) / 2
        
        semantic_score.append(score)
        
        prev_cat = cat
        
        available_cats.remove(prev_cat)

    return np.mean(semantic_score)
```

```python
def generate_null_factor_scores(num_correct, category_list, list_length=12, N=10000): 

    random_lists = np.array([np.random.choice(np.arange(1,list_length+1), num_correct, replace=False) for _ in range(N)])

    temporal_factor_score = []
    semantic_factor_score = []
    
    for rl in random_lists:
        temp_score = compute_temporal_factor_score(list(rl))
        semantic_score = compute_semantic_factor_score(list(rl), category_list)
        temporal_factor_score.append(temp_score)
        semantic_factor_score.append(semantic_score)
        
    return temporal_factor_score, semantic_factor_score
```

```python
s_i_all_subjects = []
frac_intrusions_out_of_all = []
frac_correct_out_of_all = []
list_num = dd_trials['list_num']
categories = dd_trials['category_array']
num_trials_per_subject = []
finished_subjects = []
list_length = 12
recall_position = dd_trials['position']

subject_name = []
semantic_score = []
temporal_score = []
correct_score = []
for us in subj_elec_labels_unique:
    
    subject = us.split('_')[0]
    
    if subject in finished_subjects:
        continue
        
    finished_subjects.append(subject)


    elec_idxs = np.argwhere(subj_elec_labels==us)
    num_trials_per_subject.append(elec_idxs.shape[0])
 
    clustered_e = clustered[elec_idxs]
    correct_e = correct[elec_idxs]
    list_num_e = list_num[elec_idxs]
    category_e = categories[elec_idxs]
    position_e = recall_position[elec_idxs]

    for l in range(0, list_num_e.shape[0], list_length):
        clustered_e_l = clustered_e[l:l+list_length].squeeze()
        correct_e_l = correct_e[l:l+list_length].squeeze()
        category_e_l = category_e[l:l+list_length].squeeze()
        position_e_l = position_e[l][0] # same for all list elements 
        
        # remove intrusions and duplicates
        position_e_l = [p for p in position_e_l if (p>0) & (p<13)]
        position_e_l = list(dict.fromkeys(position_e_l))
        
        correct_idxs_e_l = np.argwhere(correct_e_l==1)
        clustered_e_l_correct = clustered_e_l[correct_idxs_e_l]
        
        # if zero or 1 item is recalled, can't compute a score
        if len(position_e_l) < 2:
            continue

        category_e_l_ints = convert_cat_to_ints(category_e_l)
        semantic_score_e_l = compute_semantic_factor_score(position_e_l, category_e_l_ints)
        temporal_score_e_l = compute_temporal_factor_score(position_e_l)
        
        semantic_score.append(semantic_score_e_l)
        temporal_score.append(temporal_score_e_l)
        correct_score.append(np.sum(correct_e_l)/list_length)
        subject_name.append(subject)
        
behavorial_factor_scores_pd = pd.DataFrame({'S':semantic_score, 'TE':temporal_score, 'C': correct_score, 'SU':subject_name})
```

```python
factor_scores_avg_subj = behavorial_factor_scores_pd.groupby(['SU']).mean().reset_index()
subject_list_counts = behavorial_factor_scores_pd['SU'].value_counts()
normalized_list_count = np.array(subject_list_counts)/np.max(subject_list_counts)
```
