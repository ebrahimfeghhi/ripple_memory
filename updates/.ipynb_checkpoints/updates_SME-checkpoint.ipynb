```python
import pandas as pd
from IPython.display import display_html
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from IPython.display import display, Image    
```

```python
def plot_results(ripple_regions, region_name, hpc_region_str, ripple_str='100_1100'):
    
    img_path_ripple = f'figures/Ripple {ripple_regions} {region_name} {ripple_str}'
    img_path_no_ripple = f'figures/No ripple {ripple_regions} {region_name} {ripple_str}'
    
    mem_path = f'stats_results/mem_model_SME_{ripple_regions}_{region_name}_{ripple_str}'
    ols_path = f'stats_results/OLS_model_SME_{ripple_regions}_{region_name}_{ripple_str}'
    
    if hpc_region_str == '':
        img_path_ripple += '.png'
        img_path_no_ripple += '.png'
        mem_path += '.csv'
        ols_path += '.csv'
    else:
        img_path_ripple += f' {hpc_region_str}.png'
        img_path_no_ripple += f' {hpc_region_str}.png'
        mem_path += f'_{hpc_region_str}.csv'
        ols_path += f'_{hpc_region_str}.csv'
        
    # read images
    img_A = mpimg.imread(img_path_ripple)
    img_B = mpimg.imread(img_path_no_ripple)

    # display images
    fig, ax = plt.subplots(1,2, figsize=(20,10))
    ax[0].axis('off')
    ax[1].axis('off')
    ax[0].imshow(img_A)
    ax[1].imshow(img_B)
    plt.show()
    

    df1 = pd.read_csv(mem_path, index_col=0)
    df2 = pd.read_csv(ols_path, index_col=0)

    df1_styler = df1.style.set_precision(3).set_table_attributes("style='display:inline'").set_caption('Mixed effect model')
    df2_styler = df2.style.set_precision(3).set_table_attributes("style='display:inline'").set_caption('OLS model')

    display_html(df1_styler._repr_html_()+df2_styler._repr_html_(), raw=True)
```

```python
plot_results('ca1', 'HPC', 'ca3_dg')
# don't include ca3 
```

```python
plot_results('ca1', 'HPC', 'ca1')
```

```python
plot_results('ca1', 'AMY', '')
```

```python
plot_results('ca1', 'nonHPC_MTL', '')
```
